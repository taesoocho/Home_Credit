```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:.5f}'.format
```

```python
installments_df = pd.read_csv('installments_payments.csv')
```

```python
installments_df.info(verbose=True, null_counts=True)
```

```python
installments_df.describe(include='all')
```

```python
installments_df.columns
```

```python
pk_id = sorted(list(installments_df['SK_ID_PREV'].unique()))
```

```python
level_0 = ['NUM_INSTALMENT_VERSION', 'NUM_INSTALMENT_NUMBER', 'DAYS_INSTALMENT', 'DAYS_ENTRY_PAYMENT', 'AMT_INSTALMENT', 'AMT_PAYMENT']
level_1 = ['_MEAN', '_MEDIAN', '_STD', '_FIRST']

col_names = []
for name1 in level_0 :
    for name2 in level_1 :
        col_names.append(name1+name2)
```

```python
merge_real_num_df = pd.DataFrame(installments_df.groupby('SK_ID_PREV').agg(['mean','median','std','first']))
merge_real_num_df.reset_index(drop = True, inplace = True)
merge_real_num_df.columns = merge_real_num_df.columns.droplevel(0)
merge_real_num_df = merge_real_num_df.iloc[:,4:]
merge_real_num_df.columns = col_names
merge_real_num_df
```

```python
installments_pre = merge_real_num_df
installments_pre.index = pk_id
installments_pre = installments_pre.reset_index()
installments_pre = installments_pre.rename(columns={'index':'SK_ID_PREV'})
installments_pre
```

```python
installments_pre.to_csv('installments_merge.csv')
```
